## WF, SM, IR columns all have stars ★ in them and it need to be removed

### Cleaning the Contract column

```python
df.Contract.unique()
```

### Observaiton: this column is very messy an requires some outside the box solution
### Solution : Categorize the values into 'ON LOAN'; for players on loan, 'ACTIVE'; for players on contract and 'FREE';for free players

```python
def contract(a):
    if 'On Loan' in a:
        a = 'ON LOAN'
        return a
    elif '~' in a:
        a = 'ACTIVE'
        return a
    elif 'Free' in a:
        a = 'FREE'
        return a
```

```python
df.Contract = df.Contract.apply(contract)
```

```python
df.Contract.unique()
```

```python
df.Contract = df.Contract.astype('category')
```

```python
df = df.rename(columns={'Contract' : 'ContractStatus'})
```

## Cleaning the Hits column  

```python
df.Hits.unique()
```

### problem :some values  are written with 'k' to sgnify thaousand
### solution: change it to its actual numerical digits

```python
# function that replaces the 'k' with an empty string and cast the value to a float so it can be multiplied by 1000 and return its numerical digits

def hitChange(h):
    if 'K' in h:
        h = h.replace('K', "")
        h = float(h)* 1000
        return round(h)
    else:
        return h
```

```python
df.Hits = df.Hits.apply(hitChange)
```

```python
df.Hits.unique()
```

```python
df.head()
```

## Cleaning the height and weight column

```python
df.Height.unique()
```

```python
def convert_height(x):
    if x[-1] == '"':
        x = x.replace("\"", "")
        inch = int(x[2:]) * 2.54
        foot = int(x[0]) * 30.48
        return round(foot+inch)
    elif x[-1] == "m":
        return int(x[:-2])
```

```python
df.Height = df.Height.apply(convert_height)
```

```python
df.Height.unique()
```

```python
df = df.rename(columns={'Height':'Height_cm'})
```

```python
df.head()
```

```python
df.Weight.unique()
```

```python
def weight_kg(x):
    if 'lbs' in x:
        x = x.replace('lbs', '')
        x = float(x) * 0.45359237
        x = int(x)
        return x
    else:
        return int(x[:-2])
```

```python
df.Weight = df.Weight.apply(weight_kg)
```

```python
df.Weight.unique()
```

```python
df = df.rename(columns={'Weight':'Weight_kg'})
```
