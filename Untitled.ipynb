```python
#create copy to keep original safe
fifa_clean = fifa_df.copy()
```

```python
#remove exact duplicate rows
b4_removing_duplicates = fifa_clean.shape[0]
print('b4 removing diplicates')
b4_removing_duplicates

fifa_clean  = fifa_clean.drop_duplicates()

after_removing_duplicates = fifa_clean.shape[0]
print('after removing diplicates')
after_removing_duplicates
```

```python
#strip whitespaces from column names
new_columns = []
for col in fifa_clean.columns:
    col_clean = col.strip()
    col_clean = col_clean.replace("\u2193", "")
    new_columns.append(col_clean)
fifa_clean.columns = new_columns
```
